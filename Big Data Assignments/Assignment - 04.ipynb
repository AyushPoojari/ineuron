```

from pyhive import hive

def create_employees_table():
    connection = hive.connect(host='localhost', port=10000, database='default')
    cursor = connection.cursor()

    query = '''
    CREATE TABLE Employees (
        id INT,
        name STRING,
        salary DOUBLE
    )
    '''
    cursor.execute(query)
    connection.close()

# Example usage
create_employees_table()

from pyhive import hive

def get_customers_over_30():
    connection = hive.connect(host='localhost', port=10000, database='default')
    cursor = connection.cursor()

    query = '''
    SELECT *
    FROM Customers
    WHERE age > 30
    '''
    cursor.execute(query)
    results = cursor.fetchall()
    connection.close()
    return results

# Example usage
customers_over_30 = get_customers_over_30()
for row in customers_over_30:
    print(row)

from pyhive import hive

def sort_logs_by_timestamp():
    connection = hive.connect(host='localhost', port=10000, database='default')
    cursor = connection.cursor()

    query = '''
    SELECT *
    FROM Logs
    ORDER BY timestamp DESC
    '''
    cursor.execute(query)
    results = cursor.fetchall()
    connection.close()
    return results

# Example usage
sorted_logs = sort_logs_by_timestamp()
for row in sorted_logs:
    print(row)

from pyhive import hive

def calculate_average_salary():
    connection = hive.connect(host='localhost', port=10000, database='default')
    cursor = connection.cursor()

    query = '''
    SELECT AVG(salary) AS average_salary
    FROM Employees
    '''
    cursor.execute(query)
    results = cursor.fetchall()
    connection.close()
    return results[0][0]

# Example usage
average_salary = calculate_average_salary()
print("Average salary:", average_salary)


from pyhive import hive

def create_partitioned_sales_table():
    connection = hive.connect(host='localhost', port=10000, database='default')
    cursor = connection.cursor()

    query = '''
    CREATE TABLE Sales_Data (
        id INT,
        product STRING,
        amount DOUBLE
    )
    PARTITIONED BY (year INT, month INT)
    '''
    cursor.execute(query)
    connection.close()

# Example usage
create_partitioned_sales_table()

from pyhive import hive

def perform_inner_join():
    connection = hive.connect(host='localhost', port=10000, database='default')
    cursor = connection.cursor()

    query = '''
    SELECT *
    FROM Orders
    INNER JOIN Customers
    ON Orders.customer_id = Customers.id
    '''
    cursor.execute(query)
    results = cursor.fetchall()
    connection.close()
    return results

# Example usage
joined_data = perform_inner_join()
for row in joined_data:
    print(row)

from pyhive import hive

def process_json_logs():
    connection = hive.connect(host='localhost', port=10000, database='default')
    cursor = connection.cursor()

    query = '''
    CREATE TABLE User_Activity_Logs (
        user_id INT,
        activity STRING
    )
    ROW FORMAT SERDE 'org.openx.data.jsonserde.JsonSerDe'
    '''
    cursor.execute(query)
    connection.close()

# Example usage
process_json_logs()

```
