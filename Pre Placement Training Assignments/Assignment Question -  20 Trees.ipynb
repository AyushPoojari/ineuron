Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8
    
   / Â   / Â    \

 1Â    5  Â    10

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def constructBST(levelOrder):
    if not levelOrder:
        return None
    
    nodes = []
    root = TreeNode(levelOrder[0])
    nodes.append(root)
    
    i = 1
    while i < len(levelOrder):
        node = TreeNode(levelOrder[i])
        nodes.append(node)
        
        parent = None
        for n in nodes:
            if n.val > node.val:
                parent = n
                break
        
        if parent.left is None:
            parent.left = node
        else:
            parent.right = node
        
        i += 1
    
    return root
levelOrder = [7, 4, 12, 3, 6, 8, 1, 5, 10]
root = constructBST(levelOrder)

# Output the constructed BST using inorder traversal
def inorderTraversal(node):
    if node is None:
        return
    
    inorderTraversal(node.left)
    print(node.val, end=' ')
    inorderTraversal(node.right)

inorderTraversal(root)
```

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

```python
def canRepresentBST(arr):
    if not arr:
        return True

    root = arr[0]
    leftSubtree = []
    rightSubtree = []

    for i in range(1, len(arr)):
        if arr[i] < root:
            leftSubtree.append(arr[i])
        else:
            rightSubtree = arr[i:]
            break

    for num in rightSubtree:
        if num < root:
            return False

    return canRepresentBST(leftSubtree) and canRepresentBST(rightSubtree)


# Example usage:
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
print(canRepresentBST(arr1))  # Output: True

arr2 = [11, 6, 13, 5, 12, 10]
print(canRepresentBST(arr2))  # Output: False
```
